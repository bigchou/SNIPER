```python
import json, cv2
from matplotlib import pyplot as plt
import numpy as np
import cv2
```

```python
with open("bbox_info.json") as f:
    data = json.load(f)
```

```python
data.keys()
```

```python
data['demo/ooo/010F01867C505ED5F258E56C249714B8EB2FE87A.jpeg']
```

```python
img = cv2.imread('demo/ooo/010F01867C505ED5F258E56C249714B8EB2FE87A.jpeg')
```

```python
plt.imshow(img[:,:,::-1])
```

```python
color = (0,0,0)
for boxes in data['demo/ooo/010F01867C505ED5F258E56C249714B8EB2FE87A.jpeg']:
    y1, x1, y2, x2, conf = boxes
    cv2.rectangle(img, (x1, y1), (x2, y2), color, 5)#thickness=5
    cv2.putText(img, "%2.1f%%"%(conf*100.0), (x1, y1-10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, color, 2)
```

```python
plt.imshow(img[:,:,::-1])
```

# Draw all images

```python
results = {}
color = (0,0,0)
for src in data:
    img = cv2.imread(src)
    print('read %s'%(src))
    for boxes in data[src]:
        y1, x1, y2, x2, conf = boxes
        cv2.rectangle(img, (x1, y1), (x2, y2), color, 5)#thickness=5
        cv2.putText(img, "%2.1f%%"%(conf*100.0), (x1, y1-10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, color, 2)
    results[src] = img
for src in results:
    print("show %s"%(src))
    plt.figure()
    plt.imshow(results[src][:,:,::-1])
```
